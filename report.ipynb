```python
import nest_asyncio
nest_asyncio.apply()

tokenizer = Tokenizer()
tokenizer.remove_special_chars = False
tokenizer.max_iteration = 10000 
tokenizer.min_pair_freq = 40
tokenizer.vocab_file_path = f"vocab_{0}_{0}_special_chars_false.txt"
tokenizer.output_file_path = f"output_{0}_{0}_special_chars_false.txt"

asyncio.run(tokenizer.update_vocab())
tokenizer.tokenize()
```

```python
tokenizer = Tokenizer()

for i in range(10000, 30001, 5000):
    tokenizer.max_vocab=i
    tokenizer.vocab_file_path = f"vocab_{0}_{0}_vocab_size_{i}.txt"
    tokenizer.output_file_path = f"output_{0}_{0}_vocab_size_{i}.txt"

    asyncio.run(tokenizer.update_vocab())
    tokenizer.tokenize()
```
